```python
import pandas as pd
```

```python
microdadosEnem = pd.read_csv("microdados_enem_2019/DADOS/MICRODADOS_ENEM_2019.csv", sep=";", encoding='ISO-8859-1') 
```

```python
microdadosEnem
```

```python
microdadosEnem.columns.values
```

```python
microdadosEnemSelecionado = microdadosEnem.filter(items=['CO_MUNICIPIO_ESC','NO_MUNICIPIO_ESC',\
'CO_UF_ESC','SG_UF_ESC','NU_NOTA_CN','NU_NOTA_CH','NU_NOTA_LC','NU_NOTA_MT',])
```

```python
microdadosEnemSelecionado
```

```python
mediaEnemPorMunicipio = microdadosEnemSelecionado.groupby(['CO_MUNICIPIO_ESC','NO_MUNICIPIO_ESC',
'CO_UF_ESC','SG_UF_ESC'], as_index=False).mean()
```

```python
mediaEnemPorMunicipio
```

```python
mediaEnemPorMunicipio.to_csv("mediaEnemPorMunicipio.csv")
```
